```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Завантажуємо дані
data = pd.read_csv('IrisData_full.csv')

# Припускаємо, що остання колонка — це клас (цільова змінна), а решта — ознаки
X = data.iloc[:, :-1].values  # Ознаки (Sepal Length, Sepal Width, Petal Length, Petal Width)
y = data.iloc[:, -1].values   # Цільова змінна (Species)

print("Дані завантажено. Перші 5 рядків ознак (X):")
print(X[:5])
print("\nПерші 5 міток (y):")
print(y[:5])
```

```python
# Розділяємо дані на навчальну (70%) і тестову (30%) вибірки
# Параметр 'shuffle=True' (за замовчуванням) забезпечує перемішування записів
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print(f"\nРозмір навчальної вибірки (X_train): {X_train.shape}")
print(f"Розмір тестової вибірки (X_test): {X_test.shape}")
```

```python
scaler = StandardScaler()

# Навчаємо scaler тільки на навчальній вибірці (X_train)
X_train = scaler.fit_transform(X_train)

# Застосовуємо навчений scaler до тестової вибірки (X_test)
X_test = scaler.transform(X_test)

print("\nДані нормалізовано (стандартизовано).")
print("Перші 5 нормалізованих рядків X_train:")
print(X_train[:5])
```

```python
k_values = range(1, 21)
test_accuracies = []

for k in k_values:
    # Ініціалізація та навчання моделі k-NN
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)

    # Прогнозування та оцінка якості на тестовій вибірці
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    test_accuracies.append(accuracy)

print("\nМоделі навчено та оцінено для K від 1 до 20.")
```

```python
# Знаходимо найкраще значення K
best_k_index = np.argmax(test_accuracies)
best_k = k_values[best_k_index]
best_accuracy = test_accuracies[best_k_index]

print("-" * 50)
print(f"Найкраще значення K: {best_k}")
print(f"Максимальна точність (Accuracy) на тестовій вибірці: {best_accuracy:.4f}")
print("-" * 50)

# Візуалізація
plt.figure(figsize=(10, 6))
plt.plot(k_values, test_accuracies, marker='o', linestyle='-', color='blue')
plt.plot(best_k, best_accuracy, marker='*', markersize=12, color='red', label=f'Оптимальне K={best_k} (Accuracy={best_accuracy:.4f})')
plt.title('Залежність точності класифікації від значення K (тестова вибірка)')
plt.xlabel('Значення K (кількість сусідів)')
plt.ylabel('Точність класифікації (Accuracy)')
plt.xticks(k_values)
plt.grid(True)
plt.legend()
plt.show()
```
